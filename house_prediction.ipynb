```python
from sklearn.preprocessing import StandardScaler
df1=df.drop(['id','date', 
       'sqft_lot', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15'],axis=1)
x = df1.drop('price',axis=1)
y = df1['price']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=200,random_state=42)
model.fit(X_train,y_train)
```

```python
def pridict_price(model,scaler,user_input):
    user_input_scaled=scaler.transform([user_input])
    predict=model.predict(user_input_scaled)
    return predict[0]
```

```python
house_age=house_year(yr_build)
print("house number of year is :",house_age)
```

```python
user_input=[bedrooms, bathrooms,sqft_living,floors,yr_build]
price=pridict_price(model,scaler,user_input)
print(f"\n Current House Price: ${price:,.2f}")
```
